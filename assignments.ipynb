```python
#Write a program that asks the user to input a number and prints whether the number is positive and even, positive and odd, or negative.

n = int(input("Please enter a number: "))

if n>0:
    if n%2==0:
        print("The number is positive and even.")
    else:
        print("The number is positive and odd.")
elif n<0:
    if n%2==0:
        print("The number is negative and even.")
    else:
        print("The number is negative and odd.")
else:
    print("The number is zero.")
```

```python
##Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.
n = int(input("Please enter a number (0 to stop): "))
total = 0
while n != 0:
    total += n
    n = int(input("Please enter a number (0 to stop): "))
print("The sum of all input numbers is:", total,)
```

```python
##Write a program that prints all the numbers from 1 to 10 except 5 using a for loop and continue statement.
for i in range(1,11):
    if i == 5:
        continue
    print(i, end=' ')
```

```python
# Write a program that asks the user to input a number 
# and prints all the even numbers from 1 to that number using a for loop.

n = int(input("Please enter a number: "))
for i in range(1, n + 1):
    if i % 2 == 0:
        print(i)
```

```python
n = int(input("Please enter a number: "))

factorial = 1
numbers = []

current = n
while current >= 1:
    numbers.append(str(current))
    factorial *= current
    current -= 1

print(f"{n}! = {' × '.join(numbers)} = {factorial}")
```

```python
# Write a program that calculates the sum of the digits of a number input by the user using a while loop.
n = int(input("Please enter a number: "))
count = 0
while n > 0:
    count += 1
    n //= 10
print(f"The number of digits in the entered number is: {count}")
```

```python
n = int(input("Please enter a number: "))

if n <= 1:
    print("Please enter a number greater than 1.")
else:
    is_prime = True
    for i in range(2,n):
        if n%i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{n} is a prime number.")
    else:
        print(f"{n} is not a prime number.")
```

```python
# Write a program that prints the first n Fibonacci numbers, where n is input by the user.
n = int(input("Please enter a number: "))

a,b = 0,1

for i in range(n):
    print(a,end=' ')
    a,b = b,a+b
```
