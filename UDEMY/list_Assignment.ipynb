```python
# Create a new list containing the squares of the first 10 positive integers using a list comprehension.
# Print the new list.
num2 = [i**2 for i in range(1,11)]
print(num2)

num3 = []
for i in range(1,11):
    num3.append(i**2)
print(num3)
```

```python
# Create a new list only the even no. from list created in Assignment 1 using a list comprehension. 
# Print the new list.
num4 = [i for i in range(1, 21) if i%2 == 0]
print(num4)

num5 = []
for i in range(1,21):
    if i%2==0:
        num5.append(i)
print(num5)
```
