```python
a = int(input("Enter a number : "))

original = a
reverse = 0

while a>0:
    digit = a % 10 #a = 101, then 101%10=1, then 10%10=0, then 1%10=1
    reverse = reverse * 10 + digit #reverse = 0*10+1=1, then 1*10+0=10, then 10*10+1=101
    a=a//10 #a = 101//10=10, then 10//10=1, then 1//10=0
    
print("Palindrome" if original == reverse else "Not Palindrome")
```

```python
m = int(input("Enter a number : "))
original = m
reverse = 0

for i in range(len(str(m))): #len(str(m)) gives the number of digits in m, for example if m=101, then len(str(m))=3, so the loop will run 3 times
    reverse = reverse * 10 + (m%10) #m%10 gives the last digit of m, for example if m=101, then m%10=1, then 10%10=0, then 1%10=1
    m //= 10 #m = m//10 is the same as m = m//10, it gives the number without the last digit, for example if m=101, then m//10=10, then 10//10=1, then 1//10=0
    
print("Palindrome" if original == reverse else "Not Palindrome")
```

```python
#middle number
a = int(input("Enter a number: "))  # if a = 123
md = len(str(a))//2 #len(str(a)) gives the number of digits in a, for example if a=123, then len(str(a))=3, so m_d = 3//2 = 1
middle_digit = (a // 10**md) % 10  # a // 10**md gives the number without the last md digits, for example if a=123, then a // 10**1 = 123 // 10 = 12, then 12 % 10 = 2, so the middle digit is 2
print("Middle digit:", middle_digit)
```
