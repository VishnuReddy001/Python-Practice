**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
a = int(input("Enter any year to check leap year or not: "))
if (a%4==0 and a%100!=0) or (a%400==0):
    print(a, " is a leap year")
else:
    print(a, " is not a leap year")
```
