```python
a = int(input("Enter a number : "))

original = a
reverse = 0
while a > 0: 
    reverse = reverse * 10 + (a%10)
    a = a // 10

if original == reverse:
    print(original, " is a palindrome")
else:
    print(original, " is not a palindrome")
```

```python
num = int(input("Enter a number : "))
if num > 1:
    i = 1
    f = 0
    while i <= num:
        print(i)
        f += 1
        i += 2
    print(f"Count of odd numbers from {1} to {num}: {f}")
else:
    print("Invalid input. Please enter a number greater than 1.")
```

```python
num = int(input("Enter a number : "))
num1 = int(input("Enter another number : "))
a = num
if num > 1:
    stop = num1
    count = 0
    while num <= stop:
        print(num)
        count += 1
        num += 2
    print(f"Count of odd numbers from {a} to {num1}: {count}") # here f in print statement is used to show the variable value in the string, it is called f-string
else:
    print("Invalid input. Please enter a number greater than 1.")
```

```python
num_of_Test_cases = int(input("Enter the number of test cases : "))
total_case = 1
while total_case <= num_of_Test_cases:
    a = int(input("Enter a number : "))
    i = 1
    while i <= 10:
        print(a*i, end=' ') # print(f"{a} * {i} = {a * i}")
        i += 1
    print()  # Print a newline after each test case
    total_case += 1
```
